```python
f_v=np.where(data["12.Which network do you use more often?"]=="Airtel") 
# display(data.loc[f_v])
count_airtel=list(data.loc[f_v]['7.How long does it take you to download a song of 4 minutes?'].value_counts())
count_airtel.append(0)
print(count_airtel)
```

```python
f_v1=np.where(data["12.Which network do you use more often?"]=="Vodafone Idea (VI)") 
count_vi=list(data.loc[f_v1]['7.How long does it take you to download a song of 4 minutes?'].value_counts())
count_vi.append(0)
print(count_vi)
```

```python
f_v2=np.where(data["12.Which network do you use more often?"]=="Jio") 
count_jio=list(data.loc[f_v2]['7.How long does it take you to download a song of 4 minutes?'].value_counts())

count_jio=[int(i) for i in count_jio]
print(count_jio)
```

```python
x=np.arange(4)
width=0.2
plt.bar(x-0.2,count_airtel,width,color="cyan")
plt.bar(x,count_jio,width,color="orange")
plt.bar(x+0.2,count_vi,width,color="green")
plt.title("Time taken to download a song of 4 minutes")
plt.xticks(x,["< 1 min","1-2 mins","5-10 mins","2-5 mins"])
plt.legend(["Airtel","Jio","Vi"])
plt.show()
```

```python
s_airtel=data["17-Airtel.Rate the service quality"].dropna()
s_jio=data["16-Jio.Rate the service quality"].dropna()
s_vi=data["16-VI.Rate the service quality"].dropna()


columns=[s_airtel,s_jio,s_vi]
fig,ax=plt.subplots()
box=ax.boxplot(columns,notch=True,patch_artist=True)
plt.title("Airtel vs Jio vs Vi Service Rating Box Plot")
plt.xticks([1,2,3],["Airtel Service Rating","Jio Service Rating","Vi Service Rating"],rotation=10)
colors=['#00FFFF',"#FFA500","#00FF00"]

for patch,color in zip(box["boxes"],colors):
    patch.set_facecolor(color)

plt.show()
```
